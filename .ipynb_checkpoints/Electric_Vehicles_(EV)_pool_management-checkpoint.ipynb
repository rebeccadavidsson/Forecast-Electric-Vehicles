```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
sns.set()

from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt

from random import random
```

```python

df = pd.read_csv("charge_sessions_sampled.csv")
total_rows = len(df)
df = df.sample(int(total_rows / 10))

# Convert timestamp
df['session_time'] = df['timestamp_end'] - df['timestamp_start']
df['timestamp_start'] = pd.to_datetime(df['timestamp_start'], unit='s')
df['timestamp_end'] = pd.to_datetime(df['timestamp_end'], unit='s')
df['week_day'] = df['timestamp_start'].dt.dayofweek 
df["session_time_hours"] = df["session_time"] / 3600
df = df.assign(Date=df.timestamp_start.dt.floor('H'))

# Exclude outliers
# df = df[np.abs(df.session_time-df.session_time.mean()) <= (3*df.session_time.std())]

# Exclude outliers
df = df[df['session_time_hours'] <  50]
df = df[df['session_time_hours'] > 0.01]
df = df[df["kwh_charged"] / df["session_time_hours"] < 50]
df = df[df['kwh_charged'] < 100] 
df = df[df['kwh_charged'] > 0] 
print(len(df))
df.head(5)

```

```python
grouped = df.groupby("Date").mean()
grouped = grouped.reset_index()
grouped = grouped[np.abs(grouped.kwh_charged-grouped.kwh_charged.mean()) <= (3*grouped.kwh_charged.std())]
sample = grouped[grouped["Date"] > pd.to_datetime('2021-01-01')]
sample["moving_average"] = sample["kwh_charged"].rolling(window=7).mean() 
ax = sample.plot(x="Date", y="kwh_charged")
sample.plot(x="Date", y="moving_average", ax=ax, color="red")
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(7).mean()
    rolstd = timeseries.rolling(7).std()
    #Plot rolling statistics:
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    #Perform Dickey-Fuller test:
    dftest = adfuller(timeseries, autolag='AIC')
    print(dftest)
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)

# test_stationarity(sample["kwh_charged"])
```

```python
# AANTAL EVs
counted_df = df.groupby("Date").count()
y = counted_df['kwh_charged']
model = ARIMA(y, order=(1,1,0))
results = model.fit()

plt.rcParams["figure.figsize"] = (12,5)
plt.plot(y, alpha=0.9, label="observed");
plt.plot(results.fittedvalues[10:], color='red', alpha=0.5, label="predicted");
plt.legend();
plt.show()
```

```python
# new_samp
# split into train and test sets
sample_length = 500
X = y.sample(sample_length).values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
# walk-forward validation
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    # print('predicted=%f, expected=%f' % (yhat, obs))

mse = sqrt(mean_squared_error(test, predictions, squared=False))
print('Test MSE: %.3f' % mse)
```

```python
# plot forecasts against actual outcomes
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```

```python
# create a differenced series
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return np.array(diff)
 
# invert differenced value
def inverse_difference(history, yhat, interval=1):
    return yhat + history[-interval]

series = counted_df.copy()["kwh_charged"]
# seasonal difference
X = series.values
days_in_year = 365 * 24
differenced = difference(X, days_in_year)
# fit model
model = ARIMA(differenced, order=(7,0,1))
model_fit = model.fit()
# print summary of fit model
print(model_fit.summary())

```

```python

factor = sample_length / total_rows
plt.plot(y / factor, alpha=0.9, label="observed");
plt.plot(results.fittedvalues[10:] / factor, color='red', alpha=0.5, label="predicted");
print(np.mean(y) / factor, np.mean(results.fittedvalues[10:] / factor))
plt.legend()
plt.show()
```

```python
import datetime
future_steps = 1000
last_date = y.index[-1]

X = y.values
# extend index
new_dates = []
for t in range(future_steps):
  new_date = last_date + datetime.timedelta(hours=t)
  new_dates.append(new_date)

forecast = model_fit.forecast(steps=future_steps)

# invert the differenced forecast to something usable
history = [x for x in X]
day = 1
for yhat in forecast:
    inverted = inverse_difference(history, yhat, days_in_year)
    # print('Day %d: %f' % (day, inverted))
    history.append(inverted)
    day += 1

factor = len(X) / total_rows
print(factor)

new_index = list(series.index) + new_dates
print("Predict up until", new_index[-1])
new_hist = np.asarray(history) / np.asarray(factor)
plt.plot(new_index, new_hist, color='orange', label="future")
plt.axvline(x=pd.to_datetime('2021-04-04'))
plt.legend
plt.show()
  
```

```python
results_df = pd.DataFrame(new_hist, index=new_index)
number_of_EVs = results_df[results_df.index == pd.to_datetime('2021-04-04 14:00:00')][0]
number_of_EVs_end = results_df[results_df.index == pd.to_datetime('2021-04-04 16:00:00')][0]

print("Number of EVs predicted for this sample ", round(number_of_EVs.values[0]))

difference = round(number_of_EVs.values[0] - number_of_EVs_end.values[0])
if difference < 0:
  print("Number of EVs discharged between 14:00 and 16:00", abs(round(difference)))
else:
  print("Number of new EVs charging between 14:00 and 16:00", round(difference))
```
